## 5. Classficador em cascata de HAAR (Bom para processamento em tempo real)

```python
imagem3 = cv2.imread("imagens/px-people.jpg")
imagem3 = cv2.cvtColor(imagem3, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imagem3)
```

```python
imagem3_gray = cv2.cvtColor(imagem3, cv2.COLOR_RGB2GRAY)
```

```python
classificador = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
```

```python
# Com o 1.3 a imagem vai ser reduzida em 30% até que atinja o valor minimo do classificador
# o parametro minimo é 1.1
faces = classificador.detectMultiScale(imagem3_gray, 1.3, 5)
len(faces)
```

```python
faces[0]
```

```python
imagem3_anotada = imagem3.copy()
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(imagem3_anotada, (x,y), (x+w, y+h), (255,255,0), 2)
```

```python
# plt.figure(figsize=(20,10))
plt.imshow(imagem3_anotada)
```

```python
face_imagem = 0

for (x,y,w,h) in faces:
    face_imagem += 1
    imagem_roi = imagem3[y:y+h, x:x+w]
    imagem_roi = cv2.cvtColor(imagem_roi, cv2.COLOR_RGB2BGR)
    cv2.imwrite(f"face_{str(face_imagem)}.png", imagem_roi)
```

## 6. Classificação das imagens

```python
imagem_face_1 = cv2.imread("imagens/cropped_faces/s01_01.jpg")
imagem_face_1 = cv2.cvtColor(imagem_face_1, cv2.COLOR_BGR2RGB)

imagem_face_2 = cv2.imread("imagens/cropped_faces/s02_02.jpg")
imagem_face_2 = cv2.cvtColor(imagem_face_2, cv2.COLOR_BGR2RGB)

imagem_face_3 = cv2.imread("imagens/cropped_faces/s03_03.jpg")
imagem_face_3 = cv2.cvtColor(imagem_face_3, cv2.COLOR_BGR2RGB)

# plt.figure(figsize=(20, 10))
plt.subplot(131)
plt.title("Sujeito 01")
plt.imshow(imagem_face_1)

plt.subplot(132)
plt.title("Sujeito 02")
plt.imshow(imagem_face_2)

plt.subplot(133)
plt.title("Sujeito 03")
plt.imshow(imagem_face_3)

plt.show()
```

```python
faces_caminho = "imagens/cropped_faces/"
lista_arq_faces = [f for f in os.listdir(faces_caminho) if os.path.isfile(os.path.join(faces_caminho, f))]
```

```python
print(len(lista_arq_faces))
print(lista_arq_faces)
```

```python
faces_path_treino = "./treino/"
faces_path_teste = "./teste/"

if not os.path.exists(faces_path_treino):
    os.makedirs(faces_path_treino)

if not os.path.exists(faces_path_teste):
    os.makedirs(faces_path_teste)


for arq in lista_arq_faces:
    sujeito = arq[1:3]
    numero = arq[4:6]
    if int(numero) <= 10:
        shutil.copyfile(faces_caminho + arq, faces_path_treino + arq)
    else:
        shutil.copyfile(faces_caminho + arq, faces_path_teste + arq)
```

```python
def padronizar_imagem(imagem_caminho):
    imagem = cv2.imread(imagem_caminho, cv2.IMREAD_GRAYSCALE)
    imagem = cv2.resize(imagem, (200, 200), interpolation=cv2.INTER_LANCZOS4)
    return imagem
```

```python
list_faces_treino = [f for f in os.listdir(faces_path_treino) if os.path.isfile(os.path.join(faces_path_treino, f))]
list_faces_teste = [f for f in os.listdir(faces_path_teste) if os.path.isfile(os.path.join(faces_path_teste, f))]
```

```python
list_faces_treino[0]
```

```python
list_faces_teste[0]
```

```python
dados_treinamento, sujeitos = [], []
```

```python
for i, arq in enumerate(list_faces_treino):
    imagem_path = faces_path_treino + arq
    imagem = padronizar_imagem(imagem_path)
    dados_treinamento.append(imagem)
    sujeito = arq[1:3]
    sujeitos.append(int(sujeito))
    
```

```python
len(dados_treinamento)
```

```python
len(sujeitos)
```

```python
dados_teste, sujeitos_teste =[], []
```

```python
for i, arq in enumerate(list_faces_teste):
    imagem_path = faces_path_teste + arq
    imagem = padronizar_imagem(imagem_path)
    dados_teste.append(imagem)
    sujeito = arq[1:3]
    sujeitos_teste.append(int(sujeito))
```

```python
len(dados_teste)
```

```python
len(sujeitos_teste)
```

```python
plt.imshow(dados_treinamento[1], cmap="gray")
plt.title(sujeito[1])
```

### Algoritmos de classificacao EigenFaces

```python
sujeitos = np.asarray(sujeitos, dtype=np.int32)
sujeitos_teste = np.asarray(sujeitos_teste, dtype=np.int32)
```

```python
modelo_eingenfaces = cv2.face.EigenFaceRecognizer_create()
```

```python
modelo_eingenfaces.train(dados_treinamento, sujeitos)
```

```python
# Imagens para testar se o modelo esta acertando

# plt.figure(figsize=(20, 10))

plt.subplot(131)
plt.title("Sujeito " + str(sujeitos_teste[10]))
plt.imshow(dados_teste[10], cmap="gray")

plt.subplot(132)
plt.title("Sujeito " + str(sujeitos_teste[11]))
plt.imshow(dados_teste[11], cmap="gray")

plt.subplot(133)
plt.title("Sujeito " + str(sujeitos_teste[25]))
plt.imshow(dados_teste[25], cmap="gray")

plt.show()
```

```python
print(modelo_eingenfaces.predict(dados_teste[10]))
print(modelo_eingenfaces.predict(dados_teste[11]))
print(modelo_eingenfaces.predict(dados_teste[25]))
```
